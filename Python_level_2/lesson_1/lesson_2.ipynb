Написать программу, которая будет складывать, вычитать, умножать или делить два числа.
Числа и знак операции вводятся пользователем. После выполнения вычисления программа
не должна завершаться, а должна запрашивать новые данные для вычислений. Завершение
программы должно выполняться при вводе символа '0' в качестве знака операции. Если
пользователь вводит неверный знак (не '0', '+', '-', '*', '/'), то программа должна сообщать ему об
ошибке и снова запрашивать знак операции. Также сообщать пользователю о невозможности
деления на ноль, если он ввел 0 в качестве делителя.

```python
def func():
    def _str_term(num1, num2, oper):
        return f'{num1} {oper} {num2} = '
    operations = ['0', '+', '-', '*', '/']
    line_str = input('Введите number1 операция nymber2, для выполнения операции, для выхода из программы вместо операции введите 0 ').split(' ')
    while line_str[1] not in operations:
        print(f'Введена не известная операция {line_str[1]}, выбертие из {operations}')
        line_str[1] = input('Введите новую операцию ')
    if line_str[1] == 0:
        print('Выход из программы')
    else:
        num1 = int(line_str[0])
        num2 = int(line_str[2])
        oper = line_str[1]
        if oper == '+':
            print(_str_term(num1, num2, oper) + str(sum([num1, num2])))
            func()
            return
        if oper == '-':
            print(_str_term(num1, num2, oper) + str(num1 - num2))
            func()
            return
        if oper == '/':
            if num2 == 0:
                print('Вы ввели 0, а на ноль делить нельзя')
            else:
                print(_str_term(num1, num2, oper) + str(round(num1 / num2, 2)))
            func()
            return
        if oper == '*':
            print(_str_term(num1, num2, oper) + str(num1 * num2))
            func()
            return
func()
```

Посчитать четные и нечетные цифры введенного натурального числа. Например, если
введено число 34560, то у него 3 четные цифры (4, 6 и 0) и 2 нечетные (3 и 5).

```python
line_str = input('Введите натуральное число')

def count_odd_and_even(item, str, odd, even):
    if int(item) % 2 == 0:
        odd += 1
    else:
        even += 1
    if len(str) == 0:
        return odd, even  
    return count_odd_and_even(str[0], str[1:], odd, even)
count_odd, count_even = count_odd_and_even(line_str[0], line_str[1:], 0, 0)
print(f'Кол-во нечётных = {count_odd}, кол-во чётных {count_even}')

count_odd, count_even = 0, 0
for item in line_str:
    if int(item) % 2 == 0:
        count_odd += 1
    else:
        count_even += 1
print(f'Кол-во нечётных = {count_odd}, кол-во чётных {count_even}')
```

Сформировать из введенного числа обратное по порядку входящих в него цифр и вывести на
экран. Например, если введено число 3486, то надо вывести число 6843.

```python
line_str = input('Введите натуральное число')
temp_list = list(line_str)
temp_list.reverse()
print(f'{line_str} обратное по порядку {"".join(temp_list)}')

result = ''
for i in range(len(line_str)-1, -1, -1):
    result += line_str[i]
print(f'{line_str} обратное по порядку {result}')
```

Найти сумму n элементов следующего ряда чисел: 1 -0.5 0.25 -0.125 ...Количество элементов
(n) вводится с клавиатуры.

```python
n = int(input('Введите натуральное число'))
def count_func(curr, curr_n, n, count):
    curr_n += 1
    count += curr
    curr /= -2
    if curr_n == n:
        return count
    return count_func(curr, curr_n, n, count)

print(count_func(1, 0, n, 0))

curr = 1
count = 0

for i in range(0, n):
    count += curr
    curr /= -2
print(count)
```

Вывести на экран коды и символы таблицы ASCII, начиная с символа под номером 32 и
заканчивая 127-м включительно. Вывод выполнить в табличной форме: по десять пар
"код-символ" в каждой строке.

```python
print('-'*109)
for i in range(32, 127, 10):
    str = ''
    for i in range(i, i+10):
        if i < 128:
            str += f'|{i:3} = {chr(i):2}| '
    print(str)
    print('-'*109)
```

В программе генерируется случайное целое число от 0 до 100. Пользователь должен его
отгадать не более чем за 10 попыток. После каждой неудачной попытки должно сообщаться
больше или меньше введенное пользователем число, чем то, что загадано. Если за 10
попыток число не отгадано, то вывести загаданное число.

```python
import random

hidden_num = random.randint(0, 100)
print('Давай поиграем, я загадал число, а ты попробуй отгадать его')
for i in range(0, 10):
    cur_num = int(input('Введи число '))
    if cur_num == hidden_num:
        print(f'Молодец, ты отгадал, это число {hidden_num}')
        break
    if cur_num != hidden_num:
        if cur_num > hidden_num:
            print(f'Не угадал, оно меньше, у тебя осталось {10-i-1} попыток')
        else:
            print(f'Не угадал, оно больше, у тебя осталось {10-i-1} попыток')
    
```

Напишите программу, доказывающую или проверяющую, что для множества натуральных
чисел выполняется равенство: 1+2+...+n = n(n+1)/2, где n - любое натуральное число.

```python
n = int(input('Введите n - любое натуральное число'))
count = 0
for i in range(1, n+1):
    count += i
if count == n*(n+1)/2:
    print('Равенство доказано')
else:
    print('Условие не доказано')
```

Посчитать, сколько раз встречается определенная цифра в введенной последовательности
чисел. Количество вводимых чисел и цифра, которую необходимо посчитать, задаются
вводом с клавиатуры.

```python
line_str = input('Введите число и последовательность чисел через запятую 5 4,10,15,99,85').split(' ')
num_find = int(line_str[0])
list_num = line_str[1].split(',')
count = 0
for item in list_num:
    if num_find == int(item):
        count += 1
print(f'Количество {num_find} в {list_num} n = {count}')
```

Среди натуральных чисел, которые были введены, найти наибольшее по сумме цифр.
Вывести на экран это число и сумму его цифр.

```python
line_str = (input('Введите последовательность чисел, через пробел Например: 20 30 12 102 50')).split(' ')

def count_num_in_str(str):
    count = 0
    for item in str:
        count += int(item)
    return count

max_count = 0
max_num = None
for item in line_str:
    cur_count = count_num_in_str(item)
    if max_count < cur_count:
        max_count = cur_count
        max_num = int(item)
print(f'Максимальная сумма цифр в числе {max_num} и она равна {max_count}')
```
